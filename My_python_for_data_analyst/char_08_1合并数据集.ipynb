```python
# 合并数据集
import numpy as np
import pandas as pd
from pandas import DataFrame
from pandas import Series
```

```python
# 数据库风格的DataFrame合并
```

```python
df1 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df2 = DataFrame({'key': ['a', 'b', 'd'],
                 'data2': range(3)})
print(df1)
print(df2)
```

```python
# inner join，只要key(因为没有指定，所以就按重叠的列名来)相等的地方就排列组合，所以新的df里key等于b有3个值。
# 默认等于pd.merge(df1, df2, on='key'),on后面可以跟数组，但必须是2个DataFrame同时拥有。
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2,on='key') 
# 不指定的话默认是按照相同名字的列join
```

```python
# 
df3 = DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df4 = DataFrame({'rkey': ['a', 'b', 'd'],
                 'data2': range(3)})
print(df3)
print(df4)
pd.merge(df3, df4, left_on='lkey', right_on='rkey') # 指定merge依据的key
# 默认情况下，是按照inner方式连接的，所以c和d会被舍弃
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='outer') # 不匹配的地方自动填空值
```

```python
df1 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                 'data1': range(6)})
df2 = DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                 'data2': range(5)})  
print(df1)
print(df2)
pd.merge(df1, df2, on='key', how='left') # left join，c会出现，data2自动填充NA。
```

```python
pd.merge(df1, df2, how='inner')    # 内连接，c那一列会消失
```

```python
left = DataFrame({'key1': ['foo', 'foo', 'bar'],
                  'key2': ['one', 'two', 'one'],
                  'lval': [1, 2, 3]})
right = DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'rval': [4, 5, 6, 7]})
print(left,'\n',right)
pd.merge(left, right, on=['key1', 'key2'], how='outer')  # outer join，配不上的自动填充NA。
```

```python
pd.merge(left, right, on='key1') # key2重名如何处理？pd会自动补_x和_y。
```

```python
pd.merge(left, right, on='key1',suffixes=('_left','_right')) # 也可以指定重叠的列名后面补充啥
```

```python
# merge函数的参数
# left：       参与合并的左侧DataFrame
# right：      参与合并的右侧DataFrame
# how：        inner/outer/left/right，默认inner。
# on：         用于连接的列名。必须存在于左右两个DataFrame对象中。如果未指定，且其他连接键也未指定，
#              则以left和right列名的交集作为连接键。
# left_on：    左侧DataFrame中用作连接键的列
# right_on：   右侧DataFrame中用作连接键的列
# left_index： 将左侧的行索引用作其连接键
# right_index：将右侧的行索引用作其连接键
# sort：       根据连接键对合并后的数据进行排序，默认为True。
#              在处理大数据集时，禁用该选项可获得更好的性能。
# suffixes：   字符串值元组。用于追加到重叠列名的末尾。默认是('_x', '_y')。
# 上面都是按照列进行合并数据集，对应于sql中的join left_join,outer_join等操作
```

```python
# 下面开始研究索引上的合并
left1 = DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                   'value': range(6)})
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
print(left1,'\n',right1)    # 如果一个dataframe中的索引在另外一个dataframe的列中
```

```python
pd.merge(left1,right1,left_on='key',right_index=True)    # 左边按照key连接，右边按照索引进行内连接，取交集
```

```python
pd.merge(left1,right1,left_on='key',right_index=True).reset_index(drop=True)  # 在上面的基础上重新索引
```

```python
pd.merge(left1,right1,left_on='key',right_index=True, how='outer')   # 指定外连接方式，求并集
```

```python
# 对于层次化索引的数据
lefth = DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
                  'key2': [2000, 2001, 2002, 2001, 2002],
                 'data': np.arange(5.)})
righth = DataFrame(np.arange(12).reshape((6, 2)),
                  index=[['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'],
                         [2001, 2000, 2000, 2000, 2001, 2002]],
                  columns=['event1', 'event2'])
print(lefth,'\n',righth)
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'],right_index=True) # 以列表的形式指明用作合并键的多个列（非重复索引的数据集）
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'],right_index=True, how='outer') # 上面的外连接方式
```

```python
left2 = DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                  index=['a', 'c', 'e'],
                  columns=['Ohio', 'Nevada'])
right2 = DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                   index=['b', 'c', 'd', 'e'],
                   columns=['Missouri', 'Alabama'])
print(left2,'\n',right2)  
```

```python
pd.merge(left2, right2, left_index=True,right_index=True,how='outer')  # 同时使用从而合并双方的索引
```

```python
left1 = DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                   'value': range(6)})
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
print(left1,'\n',right1)    # 如果一个dataframe中的索引在另外一个dataframe的列中
left1.join(right1, on='key') # 指定join的基准列
```

```python
another = DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                    index=['a', 'c', 'e', 'f'],
                    columns=['New York', 'Oregon'])
print(another)
left2 = DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                  index=['a', 'c', 'e'],
                  columns=['Ohio', 'Nevada'])
right2 = DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                   index=['b', 'c', 'd', 'e'],
                   columns=['Missouri', 'Alabama'])
print(left2,'\n',right2) 

left2.join([right2, another]) # 沿着列合并,扩展列
```

```python
left2.join([right2, another], how='outer') # 多个DataFrame根据index做outer join
```

```python
s1 = Series([0,1], index=['a','b'])
s2 = Series([2,3,4], index=['c','d','e'])
s3 = Series([5,6], index=['f','g'])
print(s1,'\n',s2,'\n',s3,'\n')
```

```python
pd.concat([s1,s2,s3])   # 沿着axis=0进行concat操作，可以将值和索引粘合在一起
```

```python
# 默认情况下，是按照axis=0进行concat的，如果想要按照其他轴进行concat操作，需要将axis=1
```

```python
pd.concat([s1,s2,s3],axis=1,sort=False)
```

```python
print(s1,'\n',s3)
s4 = pd.concat([s1*5,s3])
print(s4)
```

```python
pd.concat([s1,s4], axis=1,sort=False)  
```

```python
pd.concat([s1,s4],axis=1,sort=False,join='inner')             #　内连接的方式进行横向concat操作
```

```python
# 也可以通过指定join_axes参数的方式指定要在其他轴上使用的索引
pd.concat([s1,s4],axis=1,join_axes=[['a','c','b','e']]) 
```

```python
# 但是上面方式存在的问题是参与连接的片段在结果中区分不出来
# 如果你想要在连接轴上面创建一个层次化索引，使用keys参数实现这个目的
print(s1,'\n',s1,'\n',s3)
result = pd.concat([s1,s1,s3],keys=['one','two','three']) # 默认是axis=0进行合并
print(result)
print(type(result))
result.unstack()
```

```python
pd.concat([s1, s2, s3],axis=1,keys=['one','two','three'])   #　按照axis=1进行合并的话
# keys就会成为Dataframe的列头
```

```python
df1 = DataFrame(np.arange(6).reshape(3, 2),index=['a','b','c'],columns=['one','two'])
df2 = DataFrame(5 + np.arange(4).reshape(2,2),index=['a','c'],columns=['three','four'])
print('df1: \n',df1,'df2: \n',df2)
pd.concat([df1, df2],axis=1, keys=['level1','level2'],sort=False)  # 列表keys做为层次化索引
```

```python
pd.concat({'level1':df1,'level2':df2},axis=1)  # 如果传入的不是列表而是字典，字典的键就会被当作keys选项的值
```

```python
pd.concat([df1,df2],axis=1,keys=['level1','level2'],names=['upper','lower'])  # 此外还有两个用于管理层次化索引创建方式的参数
```

```python
# 最后需要考虑一个问题：跟当前分析工作无关的DataFrame行索引
df1 = DataFrame(np.random.randn(3,4),columns=['a','b','c','d'])
df2 = DataFrame(np.random.randn(2,3),columns=['b','d','a'])
print(df1,'\n',df2)
result1 = pd.concat([df1,df2],sort=False)
print(result1)
```

```python
result2 = pd.concat([df1,df2],ignore_index=True,sort=False)   # 与上面不传ignore参数的区别是行索引有没有重新变化
print(result2)
print(result2-result1)
```

```python
# 下面研究以下合并重叠数据的问题
a = Series([np.nan,2.5,np.nan,3.5,4.5,np.nan],index=['f','e','d','c','b','a'])
b = Series(np.arange(len(a),dtype=np.float64),index=['f','e','d','c','b','a'])
print(a,'\n',b)
```

```python
b[-1] = np.nan
print(a,'\n',b)
np.where(pd.isnull(a),b,a) # 如果满足第一个条件的话执行第二个语句，否则执行第一个语句
```

```python
# Series中有一个combine_first方法，实现的是一样的功能，但是会进行数据对齐
b.combine_first(a)
```

```python
print(b[:-2],'\n')
print(a[2:])
b[:-2].combine_first(a[2:]) #   以b为基准，如果b中为空的话，补充a中的对应值
```

```python
# DataFrame中使用combine_first自然也会在列上做同样的事情
df1 = DataFrame({'a': [1.,np.nan,5.,np.nan],
                 'b': [np.nan, 2., np.nan,6.],
                 'c': range(2, 18, 4)})
df2 = DataFrame({'a': [5.,4.,np.nan,3.,7.],
                 'b': [np.nan,3.,4.,6.,8.]})
print(df1,'\n',df2)
df1.combine_first(df2)  # 咱们可以理解为：用参数对象中的数据df2为调用者对象的缺失数据打补丁
```

```python
# 下面开始研究重塑和轴向旋转的问题，用于重新排列表格型数据的基础运算
# 1.重塑层次化索引 
# 层次化索引为DataFrame数据的重排任务提供了一种具有良好一致性的方式。

# 主要功能有：stack 将数据的列旋转为行
# unstack 将数据的行旋转为列

```

```python
data = DataFrame(np.arange(6).reshape((2,3))
                 ,index=pd.Index(['Ohio','Colorado'],name='state')
                 ,columns=pd.Index(['one','two','three'],name='number'))
print(data)
```

```python
# 将dataframe展开为一个层次化索引的series
result = data.stack()
print(result,type(result))       # 使用该数据的stack方法即可将列转换为行，得到一个Series
```

```python
result.unstack() # 对一个层次化索引的Series，你可以用unstack将其重排为一个DataFrame
```

```python
# 默认情况下，unstack、stack操作的都是最内层，传入分层级别的编号或名称即可对其他级别进行
# unstack操作
print(result)
result.unstack(0)
```
